## Import Libraries

```r
library(ggplot2)
library(DataExplorer)
```

## Import Dataset

```r
df = read.csv("../input/los-angeles-crime-data-20102020/Crime_Data_from_2020_to_Present.csv")  # read csv file 
head(df, n = 5)
```

```r
names(df)
```

## View basic description of the dataset

```r
introduce(df)
```

## Plot basic description of the dataset

```r
plot_intro(df)
```

## View missing value distribution for airquality data

```r
plot_missing(df)
```

Let's get rid of the coolumns having more than 50% missing values and continue to explore. 

```r
df = subset(df, select = -c(Crm.Cd.2, Crm.Cd.3, Crm.Cd.4, Weapon.Used.Cd))
```

```r
final_df <- set_missing(df, list(0L, "unknown"))
plot_missing(final_df)
```

## Frequency distribution of all discrete variables

```r
plot_bar(final_df)
```

## Bar Plot of One Variable

```r
plot_bar(final_df$Vict.Sex)
```

## Frequency distribution of all discrete variables vs `AREA`

```r
## Left: frequency distribution of all discrete variables
plot_bar(final_df)
## Right: `price` distribution of all discrete variables
plot_bar(final_df, with = "AREA")
```

## View histogram of all continuous variables

```r
plot_histogram(final_df)
```

## View estimated density distribution of all continuous variables

```r
plot_density(final_df)
```

## View quantile-quantile plot of all continuous variables

```r
plot_qq(final_df)
```

## View quantile-quantile plot of all continuous variables by feature `AREA`

```r
plot_qq(final_df, by = "AREA")
```

## View overall correlation heatmap

Features with more than 15 categories will be ignored here.

```r
plot_correlation(na.omit(final_df), maxcat = 15L)
```

## Visualize principal component analysis

```r
pca_df <- na.omit(final_df[, c('Status','Status.Desc','Crm.Cd.1')])

plot_prcomp(pca_df, variance_cap = 0.9)
```

## View bivariate continuous distribution based on `Status`

```r
plot_boxplot(final_df, by = "Status") 
```

## Scatterplot `Status` with all other continuous features

```r
plot_scatterplot(final_df, by = "Status", sampled_rows = 1000L)
```
